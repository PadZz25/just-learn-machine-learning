```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

test = pd.read_csv(r"D:\ngoding\Machine_Learning\Dataset\house-prices-advanced-regression-techniques\test.csv")

missing_values = test.isnull().sum()

columns_to_drop = missing_values[missing_values > 1000].index
df = test.drop(columns=columns_to_drop)

missing_values_new = df.isnull().sum()
columns_to_fill = missing_values_new[missing_values_new > 0].index

numeric_to_fill = df[columns_to_fill].select_dtypes(include=['number']).columns
categorical_to_fill = df[columns_to_fill].select_dtypes(include=['object']).columns

if len(numeric_to_fill) > 0:
    df[numeric_to_fill] = df[numeric_to_fill].fillna(df[numeric_to_fill].median())

for column in categorical_to_fill:
    df[column] = df[column].fillna(df[column].mode()[0])
print("Sisa missing values:", df.isnull().sum().sum())


numeric_features_all = df.select_dtypes(include=['number']).columns

for feature in numeric_features_all[:5]:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.show()
```
